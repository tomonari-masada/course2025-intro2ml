### 等高線
* 2変数の関数の値を可視化するときに使う
$$z=f(x, y)$$

```python
delta = 1.0
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
```

```python
print(x)
print(y)
```

* 格子点を作る関数

```python
X, Y = np.meshgrid(x, y)
```

```python
print(X)
```

```python
print(Y)
```

* `X`と`Y`に同じインデックスでアクセスすると、格子点の座標が得られる。

```python
print(X[0,0], Y[0,0])
```

```python
print(X[1,0], Y[1,0])
```

```python
print(X[2,3], Y[2,3])
```

* $z = x^2 + y^2$という関数の値の等高線を描く。

```python
Z = X ** 2 + Y ** 2
```

```python
print(Z)
```

```python
plt.contour(X, Y, Z)
```
